```python

import pandas as pd

from cloudfail import cloudfail_no_tor

from hack_utils import load_dataframe
```

```python
disinfo_df = load_dataframe('../data/hackathon_domains_list.csv')
```

```python
disinfo_df
```

```python
domains = disinfo_df['Site'].tolist()
```

```python
hits = []
hitno = 0
for i in range(len(domains)): # [d for d in domains if d == '153news.net']:

    dns_records, crimeflare_records = cloudfail_no_tor(domains[i])
    hits.append([])

    if dns_records:
        for sect in ['host', 'dns', 'mx']:
            if isinstance(dns_records, dict) and sect in dns_records:
                hits[-1].append([ip['ip'] for ip in dns_records[sect]])

    if crimeflare_records:
        hits[-1].append([cr.strip() for cr in crimeflare_records])
    
    if not hits[-1]:
        print(f'\nNo hits for {domains[i]}\n')
    else:
        hitno += 1
        print(f'\nHits for {domains[i]}: {hits[-1]}\n')
    
    print(f'Found {hitno}/{i+1} hits...\n')
    
    
```
